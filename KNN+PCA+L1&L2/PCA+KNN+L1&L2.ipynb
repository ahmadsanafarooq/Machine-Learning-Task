```python
data['diagnosis'] = data['diagnosis'].map({'B': 0, 'M': 1})
data['diagnosis'].value_counts()
```

```python
x=data.drop(['id','diagnosis'],axis=1)
y=data['diagnosis']
```

```python
cor=x.corr()
masked_corr = np.where(np.eye(cor.shape[0]), cor, np.nan)
masked_corr_df = pd.DataFrame(masked_corr, index=cor.index, columns=cor.columns)
plt.figure(figsize=(10, 10))
sns.heatmap(masked_corr_df, annot=True, cmap='plasma', cbar=True, annot_kws={'size': 10})
plt.title('Masked Correlation Matrix Heatmap')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
standard=StandardScaler()
x_scaled=standard.fit_transform(x)
```

```python
pca=PCA(n_components=2)
x_pca=pca.fit_transform(x_scaled)
```

```python
x_pca
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x_pca[:, 0], y=x_pca[:, 1], hue=y, palette='viridis')
plt.title('PCA Results')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
knn=KNeighborsClassifier()
```

```python
knn.fit(x_pca,y)
```

```python
knn.score(x_pca,y)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
confusion=confusion_matrix(y,knn.predict(x_pca))
plt.figure(figsize=(8,6))
sns.heatmap(confusion,annot=True,fmt='d',cmap='viridis')
plt.xlabel('Predicted')
plt.ylabel('True')
```

```python
classification=classification_report(y,knn.predict(x_pca))
print(classification)
```

```python
accuracy_score=accuracy_score(y,knn.predict(x_pca))
print(accuracy_score)
```

```python
from sklearn.linear_model import Ridge, Lasso
```

```python
ridge=Ridge(alpha=50)
lasso=Lasso(alpha=50)
```

```python
ridge.fit(x_pca,y)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_pca[:, 0], ridge_predictions, label='Ridge Regression', color='red')
plt.plot(x_pca[:, 0], lasso_predictions, label='Lasso Regression', color='black')
plt.xlabel('Principal Component 1')
plt.ylabel('Predicted Values')
plt.title('Prediction Comparison: Ridge vs. Lasso')
plt.legend()
plt.grid(True)
plt.show()
```
