```python
import numpy as np
import pandas as pd
```

```python
ds=pd.read_csv('House Price Prediction Dataset.csv')
```

```python
ds.head()
```

```python
l_unique=ds['Location'].unique()
c_unique=ds['Condition'].unique()
g_unique=ds['Garage'].unique()
#for i,j in zip(l_unique , c_unique):
   #print("Unique Location = " , i)
   #print("Unique Condition = ",j)
print("Unique Locations = " , l_unique)
print("Unique Conditions =" , c_unique)
print("Unique Garage =" , g_unique)
```

```python
ds_dummies=pd.get_dummies(ds, columns=['Location' , 'Condition', 'Garage'] , dtype=int) # generate dummies
dummy_columns = ds_dummies.columns.difference(ds.columns) # to get dummy columns data
ds_dummies[dummy_columns] # print dummy column data
```

```python
merge_dummy_dataset=pd.concat([ds,ds_dummies[dummy_columns]] , axis='columns')
merge_dummy_dataset
```

```python
drop_location=merge_dummy_dataset.drop(['Location' , 'Location_Downtown' ,'Condition' , 'Garage' , 'Condition_Excellent' , 'Garage_No'] , axis='columns')
drop_location
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
x=drop_location.drop(['Price'] , axis='columns')
x
```

```python
y=drop_location['Price']
y
```

```python
model.fit(x,y)
```

```python
model.predict([[1990,2800,3,2,3,2000,0,1,0,0,1,0,0]])
```
